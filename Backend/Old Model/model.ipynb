```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

Datadir= "C:\\Users\\dhruv\\OneDrive\\Desktop\\Leaf Project\\Dataset\\archive (2)\\tomato\\train"
Categories =["Tomato___Bacterial_spot","Tomato___Early_blight","Tomato___healthy","Tomato___Late_blight","Tomato___Leaf_Mold","Tomato___Septoria_leaf_spot","Tomato___Spider_mites Two-spotted_spider_mite","Tomato___Target_Spot","Tomato___Tomato_mosaic_virus","Tomato___Tomato_Yellow_Leaf_Curl_Virus"]

for category in Categories:
    path=os.path.join(Datadir, category) #Here I am going to each of the folders that are specified for tomato
    for img in os.listdir(path) :
        img_array =cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array,cmap="gray")
        plt.show()
        break
    break
```

```python
print(img_array) #checking image array, keep in mind grayscale or rgb
```

```python
IMG_SIZE =100         #choosing a standard size for our image and trying to keep quality intact
new_array = cv2.resize(img_array,(IMG_SIZE, IMG_SIZE))
plt.imshow(new_array,cmap='gray')
```

```python
# Training our model starts from below
```

```python
training_data=[]
def create_training_data():
    for category in Categories:
        path=os.path.join(Datadir, category) #Here I am going to each of the folders that are specified for tomato
        class_num = Categories.index(category)
        for img in os.listdir(path) :
            try:
                img_array =cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array,(IMG_SIZE, IMG_SIZE))
                training_data.append([new_array, class_num])
            except Exception as e:
                pass
create_training_data()
```

```python
print(len(training_data))
```

```python
import random
random.shuffle(training_data)
```

```python
X=[]
Y=[]
```

```python
for features,label in training_data :
    X.append(features)
    Y.append(label)
X=np.array(X).reshape(-1, IMG_SIZE,IMG_SIZE, 1) #change 1 to rgb if end up using rgb
```

```python
import pickle
pickle_out =open("X.pickle","wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out =open("Y.pickle","wb")
pickle.dump(Y, pickle_out)
pickle_out.close()
```
